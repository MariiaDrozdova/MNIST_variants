```python
from datasets import load_from_disk

ds = load_from_disk("mnist")
ds_yellow = load_from_disk("mnist_yellow_stripes")
ds_white = load_from_disk("mnist_white_stripes")
```

```python
import matplotlib.pyplot as plt

def show_sample(dataset, idx=0):
    example = dataset[idx]

    img = example["image"]       # PIL.Image
    meta = {k: v for k, v in example.items() if k != "image"}

    # Show image
    plt.imshow(img, cmap="gray")
    plt.axis("off")
    plt.show()

    # Show metadata
    for k, v in meta.items():
        print(f"{k}: {v}")
```

```python
show_sample(ds["train"], 15)
show_sample(ds_white["train"], 15)
show_sample(ds_yellow["train"], 15)
```
